1. You are driving a little too fast, and a police officer stops you. <br/>
   Write a function to return one of 3 possible results: <br/> 
   "No ticket", "Small ticket", or "Big Ticket".
  
   - If your speed is 60 or less, the result is "No Ticket". 
   - If speed is between 61 and 80 inclusive, the result is "Small Ticket". 
   - If speed is 81 or more, the result is "Big Ticket". 

   Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday, your speed can be 5 higher in all cases.

```python
## Answer
def caught_speeding(speed, b):
    newSpeed = speed if b == False else speed-5
    if newSpeed <= 60: print("No Ticket")
    if newSpeed <= 80: print("Small Ticket")
    else: print("Big Ticket")
```

```python
caught_speeding(81,True)
```

2. 寫一個函數isPrime, 參數n, 判n是不是質數, 如果是回傳True, 不是回傳False

```python
## Answer
def isPrime(n):
    for i in range(2, n):
        if n%i==0: return False
    return True
```

```python
isPrime(7)
```

### Create an array

`numpy.array(object, dtype=None, copy=True, order='K', subok=False, ndmin=0)`

Parameters:
- **object**: An array, any object exposing the array interface, an object whose __array__ method returns an array, or any (nested) sequence.
- **dtype**: The desired data-type for the array. If not given, then the type will be determined as the minimum type required to hold the objects in the sequence.
- **copy**: If true (default), then the object is copied. Otherwise, a copy will only be made if __array__ returns a copy, if obj is a nested sequence, or if a copy is needed to satisfy any of the other requirements (dtype, order, etc.).
