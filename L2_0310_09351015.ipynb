### if, elif, else Statements 條件敘述
`if condition1:
    statement1
 elif condition2:
    statement2
 elif condition3:
    statement3
 ...
 else:
    statement(n+1)`

<br><br>程式運算: 
<br>條件1符合, 執行敘述1
<br>當條件1不符合但條件2符合時, 執行敘述2
<br>當條件1和條件2皆不符合時, 執行敘述3
    ...
    
[Note]:
1. else後要加`:`

```python
Image(filename = "img/if_else.png", width = 400, height = 600)
```

```python
if 1 < 2:
    print('True')
else:
    print('False')
```

```python
if 1 > 2:
    print('True')
else:
    print('False')
```

```python
if 1 == 2:
    print('True')
elif 3 == 3:
    print('half')
else:
    print('False')
```

```python
score = eval(input("請輸入總成績...."))
if score >= 90:
    print('A')
elif score < 90 & score >= 80:
    print('B')
elif score < 80 & score >= 70:
    print('C')
elif score < 70 & score <= 60:
    print('D')
else:
    print('E')
```

```python
print('九九乘法表......\n')
for i in range(1,10):
    print ('\n\n'+str(i)+'...\n')
    for j in range(1,10):
        num=i*j
        print(str(i)+'*'+str(j)+'='+str(num))
print('\n 執行完畢...\n')
```

### while 迴圈敘述
`while condition:
    statement1
 [else:
    statement2]`

```python
Image(filename = "img/while.png", width = 400, height = 600)
```

```python
i = 1
while (i < 5):
    print('i is: {}'.format(i))
    i += 1  
```

#### for 和 while 反覆執行直到終止條件成立才會離開迴圈

```python
i = 0
while True:
    i += 1  
    if i == 10: 
        continue         #不判斷而是回到迴圈, 執行下一個數字
    elif i == 15:
        break            #強制停止迴圈
    print('i is: {}'.format(i))
```

### functions 使用者函數
`def functionName (parameter1, parameter2, parameter3...):
    statements
    [return / return value]`

[Note]:
1. def列的最後要加`:`
2. 執行內容要縮排, 常用四個空格
3. 函式被呼叫才會執行，
4. 函式可以沒有參數或多個參數。當函式有多個參數時，參數有順序上的關係，參數的個數及順序均不能弄錯, 如果沒有依照定義函數時的參數順序填入, 填入參數時必須指明參數名稱。有預設值的參數要放在最後, 呼叫時有預設值的參數不一定要填

```python
# return, 函數執行完後回傳內容
def square(x):
    return x**2
outcome = square(2)
outcome
```

### input 函數
Allow user input <br/>
`input(prompt)`

[Note] <br/>
`prompt`: A String, representing a default message before the input.

```python
'''
Dosctring goes here
利用隨機樹函式設計遊戲'''
again = 'Y'
while (again == 'Y' or again == 'y'):
    player = eval(input('請問你出什麼拳(1是剪刀, 2是石頭, 3是布)?'))
    computer = random.randint(1,3) 
    print ('隨機數是' + str(computer))
    if (player ==1):
        if (computer ==1):
            print ('你是剪刀, 電腦是剪刀 ... 平手!')
        if (computer ==2):
            print ('你是剪刀, 電腦是石頭 ... 電腦贏!')
        if (computer ==3):
            print ('你是剪刀, 電腦是布 ... 你贏!')
    if (player ==2):
        if (computer ==1):
            print ('你是石頭, 電腦是剪刀 ... 你贏!')
        if (computer ==2):
            print ('你是石頭, 電腦是石頭 ... 平手!')
        if (computer ==3):
            print ('你是石頭, 電腦是布 ... 電腦贏!')
    if (player ==3):
        if (computer ==1):
            print ('你是布, 電腦是剪刀 ... 電腦贏!')
        if (computer ==2):
            print ('你是布, 電腦是石頭 ... 你贏!')
        if (computer ==3):
            print ('你是布, 電腦是布 ... 平手!')
    print('\n')
    again = input ('要繼續玩嗎(Y/N)?')
```

1. You are driving a little too fast, and a police officer stops you. <br/>
   Write a function to return one of 3 possible results: <br/> 
   "No ticket", "Small ticket", or "Big Ticket".
  
   - If your speed is 60 or less, the result is "No Ticket". 
   - If speed is between 61 and 80 inclusive, the result is "Small Ticket". 
   - If speed is 81 or more, the result is "Big Ticket". 

   Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday, your speed can be 5 higher in all cases.

```python
speed  =int(input('您的速度為:'))
b = eval(input('你今天生日?1為是0為否'))
if b == 1:
    speed = speed-5
if b == 0:
    speed = speed
if speed >= 81:
        print('恭喜獲得1800罰單')
elif speed >= 61:
        print('1200罰單不用找了')
else:
        print('快走妨礙我賺業績')
```

2. 寫一個函數isPrime, 參數n, 判n是不是質數, 如果是回傳True, 不是回傳False

```python
n = int(input('請輸入一個數字')) 
if(n==2):
    print("2是質數")
else:
    for i in range(2,n):
        if(n % i==0):
            print('%d不是質數!'% n)
            break
    else:
        print('%d是質數!'% n)
```
